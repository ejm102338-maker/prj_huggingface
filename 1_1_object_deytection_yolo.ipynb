```python
for result in results:
    # 1. 모든 박스 정보 가져오기
    boxes = result.boxes
    
    # 2. 클래스 ID가 0(person)인 인덱스만 찾기
    person_indices = (boxes.cls == 0).nonzero(as_tuple=True)[0]
    
    # 3. 사람에 해당하는 박스들만 따로 추출
    person_boxes = boxes[person_indices]
    
    # 확인용 출력
    for box in person_boxes:
        #if box.conf >= 0.28 :
        print(f"사람 발견! 좌표: {box.xyxy}, 신뢰도: {box.conf}")
```

```python
import cv2
for result in results:
    img = result.orig_img  # 원본 이미지
    
    for box in result.boxes:
        if int(box.cls) == 0:  # 클래스가 'person'일 때만
            # 좌표 추출 (x1, y1, x2, y2)
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = box.conf[0]
            
            # 이미지에 박스 그리기 (초록색)
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(img, f"Person {conf:.2f}", (x1, y1 - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imshow("Person Detection", img)
    cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2

# 1. 모델 로드 및 예측
# model = YOLO("models/yolo26n.pt")
# results = model.predict("image.jpg")

# 2. 이미지 읽기 및 BGR -> RGB 변환
img = cv2.imread("./images/yolo_image.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 3. 그래프 설정
fig, ax = plt.subplots(1, figsize=(12, 8))
ax.imshow(img)

# 4. 결과에서 '사람'만 골라 박스 그리기
cnt = 0
for result in results:
    boxes = result.boxes
    for box in boxes:
        if int(box.cls) == 0:  # 0번이 'person' and box.conf[0].item() > 0.5
            cnt += 1
            # 좌표 가져오기 (x1, y1, x2, y2)
            x1, y1, x2, y2 = box.xyxy[0].tolist()
            width, height = x2 - x1, y2 - y1
            conf = box.conf[0].item()

            # 사각형 패치 생성 (linewidth, edgecolor 등 설정 가능)
            rect = patches.Rectangle((x1, y1), width, height, linewidth=2, 
                                     edgecolor='blue', facecolor='none')
            ax.add_patch(rect)

            # 라벨 텍스트 추가
            plt.text(x1, y1 - 10, f'Person {conf:.2f}', color='white', 
                     fontsize=10, fontweight='bold', backgroundcolor='blue')

plt.axis('off') # 축 숨기기
plt.show()

#print(f"사람은 모두 {cnt}명 입니다")
```
